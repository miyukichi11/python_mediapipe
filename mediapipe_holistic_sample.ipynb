**動画の再生**

```python
!ffmpeg -i ./output.mp4 -vcodec vp9 ./output.webm

from IPython.display import HTML
import base64
import io

def play(file_path):
    video = io.open(file_path, 'r+b').read()
    encoded = base64.b64encode(video)
    return(HTML(data='''<video width="35%" height="35%" controls><source src="data:video/mp4;base64,{0}" type="video/mp4" /></video>'''.format(encoded.decode('ascii'))))

play('./output.webm')
```
